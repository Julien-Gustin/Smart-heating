```python
import sys
sys.path.insert(0, 'python/')

from optimize import OptimalHeatingController
from simulation import Simulation
from plot import multi_plot
from data import rooms
```

```python
control_horizon = 3
controller = OptimalHeatingController(alpha=1e-2, beta=1.2, omega=1)
simulation = Simulation(controller, control_horizon)
```

```python
datetime, Ts, Tz, Tw, Tr, Q_heat, Switch, Ta, obj_value = simulation.simulate(100)
```

```python
from ipywidgets import interact

@interact
def display_plot(room=rooms):
    room_nb = rooms.index(room)
    multi_plot(datetime, room, Ts[room_nb], Tz[room_nb], Q_heat[room_nb], Tw, Tr[room_nb], Switch[room_nb])
```
